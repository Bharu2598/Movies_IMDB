```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("/Users/bharathi/Documents/ADS-1/imdb_movie_dataset.csv")
data.columns = data.columns.str.strip()
data_2016 = data[data["Year"] == 2016]

def plot_revenue_vs_rating(data_2016):
    fig, ax = plt.subplots(figsize=(6, 6))   
    TopRating = data_2016['Rating'] > 7
    MediumRating = (data_2016['Rating'] < 7) & (data_2016['Rating'] > 4)
    LowRating = (data_2016['Rating'] < 4)
    ax.scatter(data_2016["Rating"][TopRating], data_2016["Revenue (Millions)"][TopRating], color="green", marker='o', zorder=3, s=5, label="Top Rating")
    ax.scatter(data_2016["Rating"][MediumRating], data_2016["Revenue (Millions)"][MediumRating], color="blue", marker='o', zorder=3, s=5, label="Medium Rating")
    ax.scatter(data_2016["Rating"][LowRating], data_2016["Revenue (Millions)"][LowRating], color="red", marker='o', zorder=3, s=5, label="Low Rating")
    desc_stats = data_2016['Rating'].describe()
    print("Descriptive Statistics for Rating:")
    print(desc_stats)
    ax.set_xlabel("IMDb Rating", fontsize=14)
    ax.set_ylabel("Revenue (Millions)", fontsize=14)
    ax.set_title("Revenue vs. Rating (2016 Movies)", fontsize=12)
    ax.legend()
    plt.show()
plot_revenue_vs_rating(data_2016)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("/Users/bharathi/Documents/ADS-1/imdb_movie_dataset.csv")
data.columns = data.columns.str.strip()
Comedy_genre = data[data['Genre'].str.contains('Comedy', na=False)]
Drama_genre = data[data['Genre'].str.contains('Drama', na=False)]
Horror_genre = data[data['Genre'].str.contains('Horror', na=False)]
Thriller_genre = data[data['Genre'].str.contains('Thriller', na=False)]

def plot_vote_distribution():    
    plt.figure(figsize=(10, 6))
    Comedy_genre["Votes"].plot.hist(bins=10, color="green", alpha=1, rwidth=0.6, label="Comedy")
    Drama_genre["Votes"].plot.hist(bins=10, color="black", alpha=0.5, rwidth=0.6, label="Drama")  
    Horror_genre["Votes"].plot.hist(bins=10, color="blue", alpha=0.4, rwidth=0.6, label="Horror")  
    Thriller_genre["Votes"].plot.hist(bins=10, color="red", alpha=0.5, rwidth=0.6, label="Thriller")  
    plt.legend(title="Genre")
    plt.xlabel("Genre", fontsize=14)
    plt.ylabel("Votes", fontsize=14)
    plt.title("Distribution of Votes by Genre", fontsize=16)
    std_votes = data['Votes'].std() 
    kurt_votes = data['Votes'].kurtosis()
    skew_votes = data['Votes'].skew()
    print(f"Standard Deviation of Votes: {std_votes}")
    print(f"Kurtosis of Votes: {kurt_votes}")
    print(f"Skewness of Votes: {skew_votes}")
    plt.show()
plot_vote_distribution()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("/Users/bharathi/Documents/ADS-1/imdb_movie_dataset.csv")
data.columns = data.columns.str.strip()
data_2016 = data[data["Year"] == 2016]

def Violin_data(data_2016):
 plt.figure(figsize=(20, 6))
 sns.violinplot(x='Rating', y='Metascore', data=data_2016, hue="Rating", split=True,palette="Blues")
 plt.title("Violin Plot of Metascore vs. Rating in 2016", fontsize=16)
 plt.xlabel("Rating", fontsize=14)
 plt.ylabel("Metascore", fontsize=14)
 Rat_Metascore_corr = data[['Rating', 'Metascore']].corr()
 print(f"Corelation:{Rat_Metascore_corr}")   
 plt.show()
Violin_data(data_2016)
```
